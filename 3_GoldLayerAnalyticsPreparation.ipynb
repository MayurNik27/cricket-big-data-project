```python
from pyspark.sql.functions import *
```

```python
silver_df = spark.table('workspace.default.cricket_silver_current_matches')
display(silver_df)
```

```python
gold_match_type_df= silver_df.groupBy("match_type").agg(count('*').alias("Match_Count"))
display(gold_match_type_df)
```

```python
gold_venue_df_2= silver_df.groupBy("venue").agg(count('*').alias("Total_Matches"))
display(gold_venue_df_2)
```

```python
team_1_df= silver_df.select(col("team_1").alias('team'))
team_2_df= silver_df.select(col("team_2").alias('team'))

all_team_df= team_1_df.union(team_2_df)

gold_team_df= all_team_df.groupBy('team').agg(count('*').alias("Matches_Played"))
display(gold_team_df)
```

```python
#passing query and displaying the results in spark DF

display(spark.sql("""SELECT COUNT(*) as Total_Matches,
COUNT(DISTINCT venue) as Total_Venues,
COUNT(DISTINCT match_type) as Total_Match_Types
FROM workspace.default.cricket_silver_current_matches"""))
```
